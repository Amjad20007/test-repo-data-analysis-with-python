```python
import math

def get_cost(sqft_walls, sqft_ceiling, sqft_per_gallon, cost_per_gallon):
    required_gallon= (sqft_walls + sqft_ceiling) / sqft_per_gallon
    cost_round = math.ceil(required_gallon) * cost_per_gallon
    return cost_round
```

```python
project_cost_b = get_cost(432,144,400,15)
print(f"The project will cost  {project_cost_b} Taka")
```

```python
def percentage_growth(num_users, yrs_ago):
  growth = (num_users [len(num_users)-yrs_ago]-num_users[len(num_users)-yrs_ago-1])/(num_users[len(num_users)-yrs_ago-1])*100
  return growth
```
